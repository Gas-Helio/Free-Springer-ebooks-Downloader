```python
import pandas as pd
import requests
import os
from tqdm import tqdm
```

```python
pd.set_option('display.max_columns', None)
books = pd.read_csv('Springer Free Books - Data, Stats, Math & Tech.csv')
books.head()
```

```python
folder = 'download_2/'
if not os.path.exists(folder):
    os.mkdir(folder)
```

```python
format_str = lambda x: x.replace(',', '-').replace('.', '').replace('/', ' ')
```

```python
for title, author, url, epn in tqdm(books[['Book Title', 'Author','DOI URL', 'English Package Name']].values):
    new_folder = folder + epn + '/'
    if not os.path.exists(new_folder):
        os.mkdir(new_folder)
        
    r = requests.get(url)
    new_url = r.url
    new_url = new_url.replace('/book/','/content/pdf/')
    new_url = new_url.replace('%2F','/')
    new_url = new_url + '.pdf'
    
    file_name = format_str(title) + ' - ' + format_str(author) + ' - ' + new_url.split('/')[-1]
    
    myfile = requests.get(new_url, allow_redirects=True)
    open(new_folder+file_name, 'wb').write(myfile.content)
```
